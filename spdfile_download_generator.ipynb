```python
import datetime
def dateRange(beginDate, endDate):
    dates = []
    dt = datetime.datetime.strptime(beginDate, "%Y%m%d")
    date = beginDate[:]
    while date <= endDate:
        dates.append(date)
        dt = dt + datetime.timedelta(1)
        date = dt.strftime("%Y%m%d")
    return dates
```

```python
date_start = '20150807'
date_end = '20200310'
```

```python
path_data_save = '../RawDatas/spa_data/'
```

```python
import os
fdown = []
for dt in dateRange(date_start, date_end):
    idir = path_data_save+str(dt)+'/'
    for ihour in range(0,24,3):
        buff=str(100+ihour)
        str_hour=buff[1:]+'00'

        dt_str=dt+'_'+str_hour
        fname='spd_geosfpit_'+dt_str+'.spd'

        ifile=idir+fname

        if not os.path.exists(ifile):
            weburl = 'http://pathdelay.net/spd/asc/geosfpit/' + fname + '\n'
            save = ' dir=' + idir + '\n'
            outfile = ' out=' + fname + '\n'
            fdown.append(weburl + save + outfile)
```
